para poder usar dataframe vamos a activar dos bibliotecas una de ellas es mumpy y la otra es pandas, las dos se complementa, y hacen cosas diferentes para crear dataframe existen varias posibilidades, sin embargo la mas usadas es la lectura directa desde un archivo

```python
import pandas as pd #importo pandas y le pongo el sobrenombre pd
import numpy as np #importo nunpy y le pongo el sobrenombre np
```

```python
#la primera forma de crear un dataframe es usando el metodo array, y luego guardarlo en una variable
dataframe_1= np.array([[2,4],[6,8],[10,12],[14,16]])
```

```python
#para crear el dataframe del arreglo anterior demos usar pandas
pd.DataFrame(dataframe_1)
```

```python
dataframe_2= np.array([[3,6,9],[12,15,18],[21,24,27],[30,33,36]])
pd.DataFrame(dataframe_2)
```

```python
#ahora vamos a cambiar el nombre de las columnas, esto se hace de la siguiente manera
df1=pd.DataFrame(dataframe_1,index=["fila1","fila2","fila3","fila4"],
                 columns=["columna1","columna2"])
```

```python
df1
```

```python
#primero creamos 2 listas , puedo crear cuantas listas quiera, entiendo que las listas son columnas del dataframe
ciudadcolombia=["bogota","medellin","cali","barranquilla"]
ciudadesinfo=[11000000,2500000,2200000,1300000]
```

```python
#ahora creamos los diccionarios con las listas
ciudades={"ciudades":ciudadcolombia, "poblacion":ciudadesinfo}
```

```python
#le paso el metodo de pandas
df4=pd.DataFrame(ciudades)
```

```python
#llamo al df4
df4
```

```python
#Cuatro paises, poblacion, expetativa de vida
paises_lat=["colombia","argentina","uruguay","ecuador"]
poblacion_paises=[51609000,45808747,3426260,17757000]
Expetativa_vida=[85,80,90,75]
```
